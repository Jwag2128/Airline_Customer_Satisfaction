```python
# Change null values in Arrival Delay column to zero and cast as interger
airline_cust_sat_df['Arrival Delay in Minutes'] = airline_cust_sat_df['Arrival Delay in Minutes'].fillna(0)
airline_cust_sat_df.astype({'Arrival Delay in Minutes': 'int64'}).dtypes
airline_cust_sat_df
```

```python
# Check for null values
airline_cust_sat_df.isnull().sum()
```

```python
# Rename columns
airline_cust_sat_df = airline_cust_sat_df.rename(columns={'id': 'Passenger_ID',
                                                          'Customer Type': 'Customer_Type',
                                                          'Type of Travel': 'Travel_Type',
                                                          'Class': 'Travel_Class',
                                                          'Flight Distance': 'Flight_Distance',
                                                          'Inflight wifi service': 'Inflight_WiFi',
                                                          'Departure/Arrival time convenient': 'Dep_Arr_Time_Convenient',
                                                          'Ease of Online booking': 'Online_Booking',
                                                          'Gate location': 'Gate_Location',
                                                          'Food and drink': 'Food_Drink',
                                                          'Online boarding': 'Online_Boarding',
                                                          'Seat comfort': 'Seat_Comfort',
                                                          'Inflight entertainment': 'Inflight_Entertainment',
                                                          'On-board service': 'Onboard_Service',
                                                          'Leg room service': 'Leg_Room',
                                                          'Baggage handling': 'Baggage_Handling',
                                                          'Checkin service': 'CheckIn_Service',
                                                          'Inflight service': 'Inflight_Service',
                                                          'Departure Delay in Minutes': 'Departure_Delay_Minutes',
                                                          'Arrival Delay in Minutes': 'Arrival_Delay_Minutes',
                                                          'satisfaction': 'Overall_Satisfaction'})
airline_cust_sat_df
```

```python
# Remove the word 'customer' from Customer_Type column
airline_cust_sat_df['Customer_Type'] = airline_cust_sat_df['Customer_Type'].str.replace('Customer', '')
airline_cust_sat_df['Customer_Type'] = airline_cust_sat_df['Customer_Type'].str.replace('disloyal', 'Disloyal')
airline_cust_sat_df['Travel_Type'] = airline_cust_sat_df['Travel_Type'].str.replace('Travel', '')
airline_cust_sat_df['Travel_Type'] = airline_cust_sat_df['Travel_Type'].str.replace('travel', '')
airline_cust_sat_df
```

```python
# Create a new column for 'Departure_Delay' with Yes or No values
conditions = [(airline_cust_sat_df['Departure_Delay_Minutes'] < 1), (airline_cust_sat_df['Departure_Delay_Minutes'] >= 1)]
values = ['No', 'Yes']
airline_cust_sat_df['Departure_Delay'] = np.select(conditions, values)
airline_cust_sat_df
```

```python
# Create a new column for 'Arrival_Delay' with Yes or No values
conditionsa = [(airline_cust_sat_df['Arrival_Delay_Minutes'] < 1), (airline_cust_sat_df['Arrival_Delay_Minutes'] >= 1)]
valuesa = ['No', 'Yes']
airline_cust_sat_df['Arrival_Delay'] = np.select(conditionsa, valuesa)
airline_cust_sat_df
```

```python
# Add column based on average satisfaction for scored columns
airline_cust_sat_df['Avg_Sat_Score'] = airline_cust_sat_df[['Inflight_WiFi', 'Dep_Arr_Time_Convenient',
                                                           'Online_Booking', 'Gate_Location',
                                                           'Food_Drink', 'Online_Boarding',
                                                           'Seat_Comfort', 'Inflight_Entertainment',
                                                           'Onboard_Service', 'Leg_Room',
                                                           'Baggage_Handling', 'CheckIn_Service',
                                                           'Inflight_Service', 'Cleanliness']].mean(axis=1)
airline_cust_sat_df
```

```python
# Create a new column for 'Customer_Sat' based on 'Avg_Sat_Score' values
condition = [(airline_cust_sat_df['Avg_Sat_Score'] < 2),
             (airline_cust_sat_df['Avg_Sat_Score'] >= 2) & (airline_cust_sat_df['Avg_Sat_Score'] < 4),
             (airline_cust_sat_df['Avg_Sat_Score'] >= 4)
            ]
value = ['Dissatisfied', 'Neutral', 'Satisfied']
airline_cust_sat_df['Customer_Sat'] = np.select(condition, value)
airline_cust_sat_df
```

```python
# Cleaned Data Df to CSV file
airline_cust_sat_df.to_csv('../Resources/Cust_Sat_Data.csv', index=False)
```
